# Apply

## 21. Crear una función para clasificar los vinos en base a su calidad: mayor que 7 --> Buena, mayor que 5 --> regular. Si no, mala.

```python
def cla_vinos(columna):
    if columna > 7:
        return 'Buena'
    elif columna > 5:
        return 'Regular'
    else:
        return 'Mala'
```

```python
vinos_df['Clasificacion'] = vinos_df.quality.apply(cla_vinos)
vinos_df
```

## 22. Crear una función para calcular el contenido de alcohol en términos de porcentaje.

```python
def porcen_alcohol(columna):
    porcentajes = pd.Series(columna).apply(lambda valor: round((valor * 100), 2))
    return porcentajes
```

```python
vinos_df['porcentaje_alcohol'] = vinos_df.alcohol.apply(porcen_alcohol)
vinos_df
```

## 23. Crear una función para convertir la columna "color" a mayuscula.

```python
def convertir_mayuscula(columna):
    new_columna = pd.Series(columna).str.upper()
    return new_columna
```

```python
vinos_df = vinos_df.color.apply(convertir_mayuscula)
vinos_df
```

## Joins

### 24. Concatenar dos DataFrames verticalmente:

```python
df1 = pd.DataFrame({"A": [1, 2, 3], "B": [4, 5, 6]})
df2 = pd.DataFrame({"A": [7, 8, 9], "B": [10, 11, 12]})
```

```python
concatenado = pd.concat([df1, df2])
concatenado
```

### 25. Concatenale este DataFrames horizontalmente:

```python
df3 = pd.DataFrame({"C": [13, 14, 15], "D": [16, 17, 18]})
```

```python
concatenado = pd.concat([concatenado,df3],axis= 1)
concatenado
```

```python
concatenado.reset_index(drop=True,inplace=True)
concatenado
```

### 26. Realizar una unión de dos DataFrames basada en sus índices (join)

```python
df4 = pd.DataFrame({"C": [19, 20, 21], "D": [22, 23, 24]}, index=[1, 2, 3])
```

```python
concatenado = concatenado.join(df4, lsuffix='_left', rsuffix='_right')
```

### 27. Fusionarle este DataFrame utilizando una columna en común (merge)

```python
df5 = pd.DataFrame({"A": [2, 3, 4], "E": [25, 26, 27]})
```

```python
concatenado = concatenado.merge(df5, on='A',how='left')
concatenado
```

### 28. Fusionale este DataFrame utilizando columnas diferentes en cada DataFrame (merge)

```python
df6 = pd.DataFrame({"F": [4, 5, 6], "E": [28, 29, 30]})
```

```python
concatenado = concatenado.merge(df6,left_on='B', right_on='F')
concatenado
```
